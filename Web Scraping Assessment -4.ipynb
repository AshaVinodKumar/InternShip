```python
driver=webdriver.Chrome(r"C:\Users\cjash\Desktop\chromedriver_win32\chromedriver.exe")
```

```python
url="https://www.naukri.com/recruiters/browse-all-recruiters"
```

```python
driver.get(url)
```

```python
search_bar = driver.find_element_by_xpath('//input[@class="sugInp"]').send_keys('Data science')
```

```python
Search_btn = driver.find_element_by_xpath('//button[@class="fl qsbSrch blueBtn"]').click()
```

```python
Name = []
Designation = []
Comp = []
Company = []
Skills = []
Location = []

#fetch details of Name
for n in driver.find_elements_by_xpath("//span[@class='fl ellipsis']"):
    Name.append(n.text)

#fetch details of Designation
for d in driver.find_elements_by_xpath("//span[@class='ellipsis clr']"):
    Designation.append(d.text)
    
#fetch details of Company
for c in driver.find_elements_by_xpath("//a[@class='ellipsis']"):
    Comp.append(c.text)

for e in range(0,len(Comp)-1,2):
    Company.append(Comp[e+1])

#fetch details of Skills Required
for s in driver.find_elements_by_xpath("//div[@class='hireSec highlightable']"):
    Skills.append(s.text)
    
#fetch details of Location
for l in driver.find_elements_by_xpath("//small[@class='ellipsis']"):
    Location.append(l.text)

#creating the Dataframe
Data_Science = pd.DataFrame({})
Data_Science['Name'] = Name[:48]
Data_Science['Designation'] = Designation[:48]
Data_Science['Company'] = Company[:48]
Data_Science['Skills_Hirefor'] = Skills[:48]
Data_Science['Location'] = Location[:48]

Data_Science
```
